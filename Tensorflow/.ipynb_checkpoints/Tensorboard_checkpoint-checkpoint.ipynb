```python
import tensorflow as tf
import numpy as np
```

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

add = tf.add(X,Y)
mul = tf.multiply(X,Y)
```

```python
add_hist = tf.summary.scalar('add_scalar',add)
mul_hist = tf.summary.scalar('mul_scalar',mul)

merged = tf.summary.merge_all()

#saver = tf.train.Saver()
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    #saver.restore(sess,tf.train.latest_checkpoint('./temp/'))
    
    writer = tf.summary.FileWriter('./board/test',sess.graph)
    
    for step in range(100):
        
        #saver.save(sess,'./temp %s'%(str(step)))
        
        summary = sess.run(merged,feed_dict={X:step*1.0,Y:2.0})
        writer.add_summary(summary,step)
        #(summary,global_step)
```
